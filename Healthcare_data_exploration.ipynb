```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample healthcare dataset (Replace with actual data)
data = {
    'Patient ID': range(1, 21),
    'Blood Pressure (mmHg)': [120, 135, 110, 145, 130, 125, 140, 150, 115, 138, 132, 128, 148, 118, 122, 136, 142, 126, 134, 140],
    'Sugar Level (mg/dL)': [90, 110, 85, 140, 95, 105, 130, 150, 88, 112, 100, 97, 135, 92, 89, 115, 125, 99, 108, 120],
    'Weight (kg)': [65, 78, 55, 85, 70, 72, 80, 90, 60, 75, 74, 68, 88, 59, 63, 77, 82, 66, 73, 79]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Display basic statistics
print(df.describe())

# Pairplot for overall trend analysis
sns.pairplot(df.drop(columns=['Patient ID']))
plt.show()

# Blood Pressure Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Blood Pressure (mmHg)'], bins=8, kde=True, color='blue')
plt.title('Blood Pressure Distribution')
plt.xlabel('Blood Pressure (mmHg)')
plt.ylabel('Count')
plt.show()

# Sugar Level vs. Weight Scatter Plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Weight (kg)'], y=df['Sugar Level (mg/dL)'], hue=df['Blood Pressure (mmHg)'], palette='coolwarm', size=df['Blood Pressure (mmHg)'], sizes=(20, 200))
plt.title('Sugar Level vs. Weight')
plt.xlabel('Weight (kg)')
plt.ylabel('Sugar Level (mg/dL)')
plt.show()

# Correlation Heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(df.drop(columns=['Patient ID']).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Health Parameters')
plt.show()
```
